```python
%reset -f
import numpy as np
import matplotlib.pyplot as plt
import pywt as w
```

```python
x = np.linspace(0, 1, num=2000)
cos = np.cos
# signal
sgn = cos(1.2/(0.75 - x)) +  cos(-1.2/(0.25 - x)) +  cos(0.1/(0.5 - x)) +  cos(0.1/(0.1 - x)) +  cos(0.1/(0.9 - x))
# wavelet transform ([0]: coef, [1] : freq)
scl = abs(w.cwt(sgn,np.linspace(1,200, endpoint=False),'gaus3')[0]) # 

#plot
fig, ax = plt.subplots(2,1,figsize=(11,8), gridspec_kw={'height_ratios': [1, 3]})

ax[0].plot(sgn, c='brown', lw=.7, antialiased=True)
ax[0].set_xlim([0, 2000])
ax[1].imshow(scl, cmap='inferno',aspect=130)
ax[1].set_yscale('symlog')

fig.set_facecolor('#304050')
ax[0].set_facecolor('#000003FF')
ax[1].axes.xaxis.set_ticklabels([])

plt.savefig('wavelet.png', dpi=300)
plt.show()
```
