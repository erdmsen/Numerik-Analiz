```python
%reset -f

from sympy import *
x=symbols('x')

f=(exp(x**2)*sin(x)*x)/(ln(exp(x))*cos(x))

dx=0.1
xx=1

xp1=f.subs({x:x+dx})
xm1=f.subs({x:x-dx})
xp2=f.subs({x:x+2*dx}) 
xm2=f.subs({x:x-2*dx}) 
xp3=f.subs({x:x+3*dx}) 
xm3=f.subs({x:x-3*dx}) 
xp4=f.subs({x:x+4*dx})

f1o1 = (-f + xp1) / dx
f1o2 = (-3*f + 4*xp1 - xp2) / (2*dx)
f2o1 = (f - 2*xp1 + xp2) / (dx**2)
f2o2 = (2*f - 5*xp1 + 4*xp2 - xp3) / (dx**2)
b1o1 = (-xm1 + f) / dx
b1o2 = (xm2  - 4* xm1 + 3*f) / (2*dx)
b2o1 = (xm2  - 2* xm1 + f) / (dx**2)
b2o2 = (-xm3 + 4* xm2 - 5*xm1 + 2*f) / (dx**2)
b3o1 = (-f   + 3* xp1 - 3*xp2 + xp3) / (dx**3)
b3o2 = (-5*f + 18*xp1 - 24*xp2 + 14*xp3 -3*xp4) / (2*dx**3)
c1o1 = (-xm1 + xp1) / (2*dx) 
c1o2 = ( xm2 - 8* xm1 + 8*xp1 - xp2) / (12*dx) # centered 1.türev 2.seviye
c2o1 = (-xm1 - 2* f + xp1) / (dx**2)
c2o2 = (-xm2 + 16*xm1 - 30*f + 16*xp1 - xp2) / (12*dx**2)
c3o1 = (-xm2 + 2* xm1 - 2*xp1 + xp2) / (2*dx**3)
c3o2 = ( xm3 - 8* xm2 + 13*xm1 - 13*xp1 + 8*xp2 - xp3) / (8*dx**3)


d1f=diff(f).subs({x:xx}).evalf().round(3)
d2f=diff(d1f).subs({x:xx}).evalf().round(3)

#gerçek değerler
d1f,d2f
```
