```python
%reset -f

import sympy as sp
import numpy as np

X,Y=sp.symbols('X Y')
```

```python
x=np.array([-1,0,2])
y=np.array([0,2,24])

#değer
```

```python
n=y.shape[0]

L = [1]*n
for j in range(n):
    for i in range(n):
        if i != j:  L[j] *= (X-x[i])/(x[j]-x[i])
# istediğin hergangi bir L terimini göster
l=3
sp.simplify(L[l-1])  
```

```python
p=0
for i in range(n): p += (L[i]*Y).subs({Y:y[i]})
p=np.around(np.array(sp.Poly(p,X).all_coeffs()).astype(np.float),4)
p=sp.Poly(p,X).as_expr()
p
#polinoma değer ver:
p.subs({X:3}).round(3)
```

```python
dp=sp.diff(p)
dp
```

```python
dp.subs({X:3})
```

```python
if sp.diff(dp)>0: print('minimum @ {}'.format(sp.solve(dp,X)[0].evalf().round(3)))
else: print('maximum @ {}'.format(sp.solve(dp,X)[0].evalf().round(3)))
```

```python
%reset -f
import math as mt
```

```python
# fonksiyon
''' bağımlı değişken i dikkat! '''
def f(i): return (i+2)/(i**2+2)
# aralık
I=(1,8)
# adım sayısı
n=6
# yöntem: 'dd' // 'ymk' // 'on'
ynt='ymk'
```

```python
# x değişim miktarı
dx=round((I[1]-I[0])/n,3)
dx
```

```python
# x - tanım kümesi dizisi
x=[]
x.clear()
x.append(I[0])
for i in range(1,n): x.append(round(x[i-1]+dx,3))
x.append(I[1])
print('x=',x)
if ynt=='on':
    x2=[]
    x2.clear()
    d2=dx/2
    for i in range(n): x2.append(round(x[i]+d2,3))
    print('x2=',x2)
# y - görüntü kümesi
y=[]
y.clear()
if ynt=='on':
    for i in range(n): y.append(round(f(x2[i]),3))
else: 
    for i in range(n+1): y.append(round(f(x[i]),3))
print('y=',y)
```

```python
# integral hesabı
integral=0
if ynt=='ymk':
    for i in range(1,n): integral+=y[i]
    integral += (y[-1]+y[0])/2
    integral *= dx
else: 
    for i in range(len(y)): integral+=y[i]
    integral *= dx    
round(integral,3)
```

<font size=1> Alan sayısı ( $2n$ ) çift olmalı

```python
%reset -f
import math as mt
```

```python
''''''
# fonksiyon
def f(x): return x
# aralık
I=[-3,5]
# çift alan sayısı
n=1
'''''';
```

```python
# x değerleri dizisi
for i in range(2*n-1):
    if i==0: dx = (I[1]-I[0])/(2*n)
    I.insert(i+1,round(I[i]+dx,3))
    i+=1
I    
```

```python
# n ayrı alan için sonuç dizisi
out=[]
out.clear()
for i in range(n):
    di = I[i+2]-I[i]
    itg= di*(f(I[i])+4*f(I[i+1])+f(I[i+2]))/6
    out.append(abs(itg))
    i+=1
print(out)
```

```python
# sonuç
integral=sum(out)
integral
```

```python
%reset -f
import math as mt
```

```python
''''''
# fonksiyon
def f(x): return x
# aralık
I=[0,6]
# çift alan sayısı
n=1
'''''';
```

```python
# x değerleri dizisi
for i in range(3*n-1):
    if i==0: dx=(I[1]-I[0])/(3*n)
    I.insert(i+1,I[i]+dx)
    i+=1
I    
```

```python
# n ayrı alan için sonuç dizisi
out=[]
out.clear()
for i in range(n):
    di = I[i+3]-I[i]
    itg= di*(f(I[i])+3*f(I[i+1])+3*f(I[i+2])+f(I[i+3]))/8
    out.append(abs(itg))
    i+=2
print(out)
```

```python
# sonuç
integral=sum(out)
integral
```
